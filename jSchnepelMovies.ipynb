```python
# Numberized all categories so we can create another visualization for all aspects
for col_name in df.columns:
    if(df[col_name].dtype == 'object'):
        df[col_name]= df[col_name].astype('category')
        df[col_name] = df[col_name].cat.codes
```

```python
df
```

```python
# Visualization matrix for all aspects
# Anything above a .5 shows a high correlation

correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Here we are unstacking the correlation matrix to create sorting pairs in the next step

correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
# Doing a quicksort to find the highest correlations

sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# From that, we can see the highest correlations are between how much a company is willing to spend on a budget
# and that the country and budget also have a high correlation
```

```python
# Look at correlation (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```
